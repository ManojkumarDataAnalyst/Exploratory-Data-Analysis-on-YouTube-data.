## Converting the published_at column from object to datetime format


```python
df_yt['published_at'] = pd.to_datetime(df_yt['published_at'])
df_yt.dtypes
```

## Creating the new column as Published_month by exracting the month from the Published_at by creating this we can gain some additional information on month-wise.

```python
df_yt['published_month'] = df_yt['published_at'].dt.strftime('%B')
df_yt['published_month']
```

```python
df_yt['published_at'].sort_values()
```

# Inference:
- This dataset contains Youtube data between 27,July2020 - 08,December2021.

##  The total number of videos published each month and arrange the month from high to low based on video count

```python
#grouping by months

video_month = df_yt.groupby(('published_month'))
```

```python
#apply aggregate function onto grouped month

video_month['title'].count().sort_values(ascending=False)
```

# Inference :
- In october month ,the number of video published is 4976 which is higher compared to other month

- In may month , the number of video published is 2073 which is lower compared to other month may be some of the youtuber might be enjoying there summer vacation.

##  The top10 channel names that having the highest number of videos 

```python
# The top 10 channel names having the highest number of videos

df_yt.groupby(by='channel_title')[['video_id']].count().sort_values(by='video_id',ascending=False).head(10)
```

```python
# The bottom 10 channel names having the highest number of videos.

df_yt.groupby(by='channel_title')[['video_id']].count().sort_values(by='video_id',ascending=False).tail(10)
```

##  The title of the video which has the maximum number of likes and the title of the video having minimum likes .

```python
Max_likes = df_yt['likes'].max()
df_yt[df_yt['likes'] == Max_likes][['title','likes']]
```

```python
# Got minimum likes
Min_likes = df_yt['likes'].min()
df_yt[df_yt['likes'] == Min_likes][['title','likes']]
```

##  The title of the video which has the maximum number of dislikes and the title of the video having minimum dislikes .


```python
# Got maximum dislikes
Max_dislikes = df_yt['dislikes'].max()
df_yt[df_yt['dislikes']==Max_dislikes][['title','dislikes']]
```

```python
# Got minimum dislikes
Min_dislikes = df_yt['dislikes'].min()
df_yt[df_yt['dislikes'] == Min_dislikes][['title','dislikes']]
```

## Inference:
- The sports channel are posting more number of videos.
- The Bts got more number of likes and views ,it a  independent music channel .
- The cuties movie trailer by netflix got more number of dislikes ,because the descripition of this may affect many people.

##  The number of views have  effect on how many people disliked the video

```python
#First we can create the correlation relationship with this two variable View count and Dislikes.
cor_twovar = df_yt['view_count'].corr(df_yt['dislikes'])
cor_twovar
```

```python
plt.figure(figsize=(7,4))
plt.title("Relationship Between View and Dislikes")
sns.scatterplot(data=df_yt,x=df_yt['view_count'],y=df_yt['dislikes'],)
plt.show()
```

## Inference :
- The correlation between views & dislikes is positive (0.684), the scatter plot shows that if views increses there is slight increases in dislikes.
